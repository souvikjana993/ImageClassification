```
from keras.layers import Dense
from keras.models import K,Model,Input
from keras.optimizers import SGD
```

```
import numpy as np
```

```
X=np.array(([0,0],[0,1],[1,0],[1,1]),dtype=np.float32)
```

```
#OR
Y=np.array(([0,1,1,1,]),dtype=np.float32)

#2d array
Y_2=np.array(([1,0],[0,1],[0,1],[0,1]),dtype=np.float32)

#AND
Z=np.array(([0,0,0,1,]),dtype=np.float32)
```

```
input_layer = Input(shape=[2],dtype=np.float32)
output_layer = Dense(units=2,activation='linear')(input_layer)

model=Model(input_layer,output_layer)
model.summary()
```

```
sgd_opt = SGD(lr=0.1)
model.compile(optimizer=sgd_opt,loss='binary_crossentropy')
```

```
model.fit(x=X,y=Y_2,epochs=200)
```

```
model.predict(x=X)
```

```
import pandas as pd
df = pd.read_csv('auto.csv',sep='\s+',header=None,na_values='?')
```

```
df = df.dropna()
df.head()
```

```
df=df.iloc[:,:-1]
df.head()
```

```
df=df.apply(pd.to_numeric)
```

```
response = df.iloc[:,0].values
predictors =df.iloc[:,2:6].values
```

```
from sklearn.preprocessing import StandardScaler
std=StandardScaler().fit_transform(predictors)
```

```
input_layer = Input(shape=[4],dtype=np.float32)
output_layer = Dense(units=1,activation='linear')(input_layer)

model=Model(input_layer,output_layer)
model.summary()
```

```
sgd_opt = SGD(lr=0.01)
model.compile(optimizer=sgd_opt,loss='mean_squared_error')
```

```
model.fit(x=std,y=response,batch_size=1,epochs=200)
```
